# TASK 1

```python
import pandas as pd
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
```

```python
url = "https://www.daftlogic.com/information-appliance-power-consumption.htm"
headers = {'User-Agent': 'Mozilla/5.0'}
```

```python
req = Request(url=url, headers=headers)
page = urlopen(req).read()
```

```python
soup = BeautifulSoup(page, 'html.parser')
soup
```

```python
table = soup.find('table', id="tblApp") ## <table> element with the id attribute set to 'tblApp (found by inspecting)
table
```

```python
if table:
    table_text = table.get_text(separator=' ')
    print(table_text)
```

```python
df = pd.read_html(str(table))[0] ## [0] is used to select the first DataFrame from the list of dataframes incase of more than one table
df
```

```python
df.to_csv("powerConsumption.csv")
```

```python
import numpy as np
import matplotlib as plt
```

```python
data = pd.read_csv('Iris.csv')
data
```

```python
# Question1: What is the average sepal length for each species of iris?
data.groupby('Species').SepalLengthCm.mean()
```

```python
# Question2: What is the correlation between sepal length and petal length?
correlation = data['SepalLengthCm'].corr(data['PetalLengthCm'])
print("Correlation between petal length and sepal length:", correlation)
```

```python
# Question3: Which species of iris has the largest petals?
max = data.groupby('Species').PetalLengthCm.max()
largest_species = largest_species = max.idxmax()
print("Species with lasrgest petals: ", largest_species)
```
